```python
import pandas as pd
from pprint import pprint
from sklearn.svm import SVC
import json
from xgboost import XGBClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from training_methods import get_results
```

### Random Forest

```python
test_variations = ['test_experiments', 'test_random_negatives']
all_results = dict()

for model in [lambda: SVC(probability=True), RandomForestClassifier, XGBClassifier]:
    all_results = dict()
    for test_name in test_variations:
        variation_results = []
        print(test_name)
        for k in [3, 4, 5]:
            positive = pd.read_csv(f'./train_data/positive_train_{k}.csv')
            negative_name = 'negative_train' if test_name == 'test_experiments' else 'negative_random_train'
            negative = pd.read_csv(f'./train_data/{negative_name}_{k}.csv')

            test_data1 = pd.read_csv(f'./test_data/test_experiments_{k}.csv')
            test_data2 = pd.read_csv(f'./test_data/test_random_negatives_{k}.csv')
            
            train_data = pd.concat([positive, negative])
            results = get_results(train_data, test_data1, test_data2, model)
            
            single_record = {
                'k': k,
                'negative_data': 'experiment' if  test_name == 'test_experiments' else 'random',
                'results': results
            }
            
            pprint(single_record)
            variation_results.append(single_record)
            
        all_results[test_name] = variation_results
    
    with open(f"results/results0_{model.__name__}.json", "w") as outfile: 
        json.dump(all_results, outfile, indent=4)
```
